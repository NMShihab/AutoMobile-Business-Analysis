# Import packages

```python
import yfinance
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

```

```python
# Set plotting style
plt.style.use('fivethirtyeight')
```

## Import data

```python
raw_data = yfinance.download (tickers = "VOW3.DE, PAH3.DE, BMW.DE", interval = "1d", group_by = 'ticker',
                              auto_adjust = True, treads = True)
```

```python
# 
raw_data
```

```python
df = raw_data.copy()
```

## Set key dates

```python
# Starting Date
start_date = "2009-04-05" 

# First Official Announcement - 49.9%
anouncement_1 = "2009-12-09" 

# Second Official Announcement - 51.1%
anouncement_2 = "2012-07-05" 

#Ending Date
end_date = "2014-01-01"

# Dieselgate
d_gate = '2015-09-20' 
```

## Preprocessing data

```python
# Extracting Closing Prices
df['vol'] = df['VOW3.DE'].Close
df['por'] = df['PAH3.DE'].Close
df['bmw'] = df['BMW.DE'].Close
```

```python
# Creating Returns
df['ret_vol'] = df['vol'].pct_change(1).mul(100)
df['ret_por'] = df['por'].pct_change(1).mul(100)
df['ret_bmw'] = df['bmw'].pct_change(1).mul(100)
```

```python
# Creating Squared Returns
df['sq_vol'] = df.ret_vol.mul(df.ret_vol)
df['sq_por'] = df.ret_por.mul(df.ret_por)
df['sq_bmw'] = df.ret_bmw.mul(df.ret_bmw)
```

```python
# Extracting Volume
df['q_vol'] = df['VOW3.DE'].Volume
df['q_por'] = df['PAH3.DE'].Volume
df['q_bmw'] = df['BMW.DE'].Volume
```

```python
# Assigning the Frequency and Filling NA Values
df = df.asfreq('b')
df = df.fillna(method='bfill')
```

```python
# Removing Surplus Data
del df['VOW3.DE']
del df['PAH3.DE']
del df['BMW.DE']
```

## Visualize data

```python
# visualize data during Start to end period
plt.figure(figsize=(20,8))
df[start_date:].vol.plot(color="Blue")
df[start_date:].bmw.plot(color="green")
df[start_date:].por.plot(color="red")
plt.show()
```

```python
# Visualize in different interval
plt.figure(figsize=(20,8))
df['vol'][start_date:anouncement_1].plot(color = "#4169e1")
df['por'][start_date:anouncement_1].plot(color = "#32cd32")
df['bmw'][start_date:anouncement_1].plot(color = "#800000")

df['vol'][anouncement_1:anouncement_2].plot(color = "#000080")
df['por'][anouncement_1:anouncement_2].plot(color = "#228b22")
df['bmw'][anouncement_1:anouncement_2].plot(color = "#960018")

df['vol'][anouncement_2:end_date].plot(color = "#003366")
df['por'][anouncement_2:end_date].plot(color = "#004953")
df['bmw'][anouncement_2:end_date].plot(color = "#C21807")

plt.legend(['Volkswagen','Porsche','BMW'])

plt.show()
```
