# Import packages

```python
import yfinance
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

```

```python
# Set plotting style
plt.style.use('fivethirtyeight')
```

## Import data

```python
raw_data = yfinance.download (tickers = "VOW3.DE, PAH3.DE, BMW.DE", interval = "1d", group_by = 'ticker',
                              auto_adjust = True, treads = True)
```

```python
# 
raw_data
```

```python
df = raw_data.copy()
```

## Set key dates

```python
# Starting Date
start_date = "2009-04-05" 

# First Official Announcement - 49.9%
anouncement_1 = "2009-12-09" 

# Second Official Announcement - 51.1%
anouncement_2 = "2012-07-05" 

#Ending Date
end_date = "2014-01-01"

# Dieselgate
d_gate = '2015-09-20' 
```

## Preprocessing data

```python
# Extracting Closing Prices
df['vol'] = df['VOW3.DE'].Close
df['por'] = df['PAH3.DE'].Close
df['bmw'] = df['BMW.DE'].Close
```

```python
# Creating Returns
df['ret_vol'] = df['vol'].pct_change(1).mul(100)
df['ret_por'] = df['por'].pct_change(1).mul(100)
df['ret_bmw'] = df['bmw'].pct_change(1).mul(100)
```

```python
# Creating Squared Returns
df['sq_vol'] = df.ret_vol.mul(df.ret_vol)
df['sq_por'] = df.ret_por.mul(df.ret_por)
df['sq_bmw'] = df.ret_bmw.mul(df.ret_bmw)
```

```python
# Extracting Volume
df['q_vol'] = df['VOW3.DE'].Volume
df['q_por'] = df['PAH3.DE'].Volume
df['q_bmw'] = df['BMW.DE'].Volume
```

```python
# Assigning the Frequency and Filling NA Values
df = df.asfreq('b')
df = df.fillna(method='bfill')
```

```python
# Removing Surplus Data
del df['VOW3.DE']
del df['PAH3.DE']
del df['BMW.DE']
```

# Best fit Model

### For Volkswagen

```python
# Import package 
from pmdarima.arima import auto_arima
```

```python
# Model for pre anouncement without exog
mod_pr_pre_vol = auto_arima(df.vol[start_date:anouncement_1], m = 5, max_p = 5, max_q = 5)
# Model for between anouncement without exog
mod_pr_btn_vol = auto_arima(df.vol[anouncement_1:anouncement_2],m = 5, max_p = 5, max_q = 5)
# Model for post anouncement without exog
mod_pr_post_vol = auto_arima(df.vol[anouncement_2:end_date],m = 5, max_p = 5, max_q = 5)
```

```python
# Model for pre anouncement
mod_pr_pre_vol_x = auto_arima(df.vol[start_date:anouncement_1], exogenous = df[['por','bmw']][start_date:anouncement_1],
                            m = 5, max_p = 5, max_q = 5)
# Model for between anouncement
mod_pr_btn_vol_x = auto_arima(df.vol[anouncement_1:anouncement_2], exogenous = df[['por','bmw']][anouncement_1:anouncement_2],
                            m = 5, max_p = 5, max_q = 5)
# Model for post anouncement
mod_pr_post_vol_x = auto_arima(df.vol[anouncement_2:end_date], exogenous = df[['por','bmw']][anouncement_2:end_date],
                            m = 5, max_p = 5, max_q = 5)
```

```python
df_auto_pred_btn_vol = pd.DataFrame(mod_pr_pre_vol.predict(n_periods = len(df[anouncement_1:anouncement_2])),index = df[anouncement_1:anouncement_2].index)
df_auto_pred_btn_vol.plot(figsize = (20,5), color = "red")

df.vol[anouncement_1:anouncement_2].plot(color = "blue")
plt.title("VW Predictions vs Real Data from "+anouncement_1+" to "+anouncement_2, size = 24)
plt.legend(['Predicted','Real'])
plt.show()
```

```python
df_auto_pred_btn_vol_x = pd.DataFrame(mod_pr_pre_vol_x.predict(n_periods = len(df[anouncement_1:anouncement_2]),
                                                               exogenous = df[['por','bmw']][anouncement_1:anouncement_2]),
                                      index = df[anouncement_1:anouncement_2].index)
df_auto_pred_btn_vol_x.plot(figsize = (20,5), color = "red")

df.vol[anouncement_1:anouncement_2].plot(color = "blue")
plt.title("VW Predictions(with Exog) vs Real Data from "+anouncement_1+" to "+anouncement_2, size = 24)
plt.legend(['Predicted','Real'])
plt.show()
```

```python
df_auto_pred_post_vol = pd.DataFrame(mod_pr_btn_vol.predict(n_periods = len(df[anouncement_2:end_date])),
                                     index = df[anouncement_2:end_date].index)
df_auto_pred_post_vol.plot(figsize = (20,5), color = "red")

df.vol[anouncement_2:end_date].plot(color = "blue")
plt.title("VW Predictions vs Real Data from "+anouncement_2+" to "+end_date, size = 24)
plt.legend(['Predicted','Real'])
plt.show()
```

```python
df_auto_pred_post_vol_x = pd.DataFrame(mod_pr_btn_vol_x.predict(n_periods = len(df[anouncement_2:end_date]),
                                                               exogenous = df[['por','bmw']][anouncement_2:end_date]),
                                      index = df[anouncement_2:end_date].index)
df_auto_pred_post_vol_x.plot(figsize = (20,5), color = "red")

df.vol[anouncement_2:end_date].plot(color = "blue")
plt.title("VW Predictions(with Exog) vs Real Data from "+anouncement_2+" to "+end_date, size = 24)
plt.legend(['Predicted','Real'])
plt.show()
```

```python
df_auto_pred_last_vol = pd.DataFrame(mod_pr_post_vol.predict(n_periods = len(df[end_date:])),
                                     index = df[end_date:].index)
df_auto_pred_last_vol.plot(figsize = (20,5), color = "red")

df.vol[end_date:].plot(color = "blue")
plt.title("VW Predictions vs Real Data from "+end_date+" to "+str(df.index[-1]), size = 24)
plt.legend(['Predicted','Real'])
plt.show()
```

```python
df_auto_pred_last_vol_x = pd.DataFrame(mod_pr_post_vol_x.predict(n_periods = len(df[end_date:]),
                                                               exogenous = df[['por','bmw']][end_date:]),
                                      index = df[end_date:].index)
df_auto_pred_last_vol_x.plot(figsize = (20,5), color = "red")

df.vol[end_date:].plot(color = "blue")
plt.title("VW Predictions(with Exog) vs Real Data from "+end_date+" to "+str(df.index[-1]), size = 24)
plt.legend(['Predicted','Real'])
plt.show()
```
